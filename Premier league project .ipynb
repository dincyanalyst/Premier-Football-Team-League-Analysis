```python
df=df[df['Matches Played']<900].reset_index(drop=True)
```

```python
df['WinningRate']=(df['Win']/df['Matches Played'])*100
df['LosingRate']=(df['Loss']/df['Matches Played'])*100
df['Drawn Rate']=(df['Drawn']/df['Matches Played'])*100
df['CLean sheet rate']=(df['Clean Sheets']/df['Matches Played'])*100
```

```python
df['Avg Goals Per match']=df['Goals']/df['Matches Played']

df['Avg Goals Per match']=df['Avg Goals Per match'].round()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))

boxplot=plt.boxplot([df['WinningRate'],df['LosingRate'],df['Drawn Rate'],df['CLean sheet rate']],
                    patch_artist=True,
                   labels=['WinningRate','LosingRate','Drawn Rate','CLean sheet rate'])

plt.title=('Distribution of WinningRate,LosingRate,Drawn Rate,CLean sheet rate' )
plt.xlabel=('WinningRate,LosingRate,Drawn Rate and CLean sheet rate')
plt.ylabel=('Rate')

plt.show()
```

```python
# Calculating interquartile range
Q1=df['WinningRate'].quantile(0.25)
Q3=df['WinningRate'].quantile(0.75)
IQR=Q3-Q1

#Calculating upper bound
upper_bound=Q3 + 1.5*IQR

#Identify Clubs with high winning rate
highwinningrate=df[(df['WinningRate']>upper_bound)]
highwinningrate
```

```python
#LowWinningrate

lower_bound=Q1-1.5*IQR

LowWinningrate=df[(df['WinningRate']<lower_bound)]
LowWinningrate
```

```python
##Drawn Rate

Q1=df['Drawn Rate'].quantile(0.25)
Q3=df['Drawn Rate'].quantile(0.75)
IQR=Q3-Q1

#Calculating upper bound for drawn rate
upper_bound=Q3 + 1.5*IQR

#Identify Clubs with high winning rate
highwinningrate=df[(df['Drawn Rate']>upper_bound)]
highwinningrate
```

```python
#Calculate the upper bound for the "Winning Rate" column
upper_bound_winningrate=df['WinningRate'].quantile(0.75)


#Calculate the lower bound for the "Losing Rate" column
lower_bound_Losingrate=df['LosingRate'].quantile(0.25)


#Calculate the lower bound for the "Drawn Rate" column
lower_bound_Drawnrate=df['Drawn Rate'].quantile(0.25)

#Calculate the upper bound for the "Clean sheet Rate" column
upper_bound_Cleansheetrate=df['CLean sheet rate'].quantile(0.75)
```

```python
len(df)
```

```python
df['scores']=np.zeros(len(df))
```

```python
df.head()
```

```python
df.loc[df['Matches Played']>=372,'scores']+=10
df.loc[df['WinningRate']>=upper_bound_winningrate,'scores']+=15
df.loc[df['LosingRate']<=lower_bound_Losingrate,'scores']+=15
df.loc[(df['Drawn Rate']<=lower_bound_Drawnrate) & (df['LosingRate']<=lower_bound_Losingrate),'scores']+=10
df.loc[(df['CLean sheet rate']>=upper_bound_Cleansheetrate) & (df['WinningRate']>=upper_bound_winningrate),'scores']+=10
df.loc[df['Winners']==1,'scores']+=15
df.loc[df['Runners-up']==1,'scores']+=10
df.loc[df['lastplayed_pl']==2023,'scores']+=15
```

```python
df_sort=df.sort_values(by='scores',ascending=False)

#create a bar chart of team scores
plt.figure(figsize=(25,10))
plt.bar(df_sort['Club'],df_sort['scores'],color='blue')

#addlabels and title to the chart
#plt.ylabel('Scores', fontsize=14)
#plt.title('Football club v/s Performance score', fontsize=18)


#adding legend to explain blue bars
plt.legend(['scores'],fontsize=14)

#rotate the team names on x-axis for better readability
plt.xticks(rotation=90,fontsize=14)
plt.yticks(fontsize=14)

#set the y-axis limit to start from 0 and end at 100
plt.ylim(0,100)

#display the chart
plt.show()
```
